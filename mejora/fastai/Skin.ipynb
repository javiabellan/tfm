```python
from fastai.vision import *
```

# Data
https://forums.fast.ai/t/load-train-and-validation-set-from-different-csv-files/35430

```python
path          = pathlib.Path("D:/Datasets/TFM/ISIC-2017")
df_train      = pd.read_csv(path / "ground_truth_train.csv")
df_valid      = pd.read_csv(path / "ground_truth_valid.csv")
df_train_meta = pd.read_csv(path / "data_train" / "ISIC-2017_Training_Data_metadata.csv")
df_valid_meta = pd.read_csv(path / "data_valid" / "ISIC-2017_Validation_Data_metadata.csv")

df_train["image_id"] = "data_train/"+df_train["image_id"]
df_train["valid"] = False
df_train["age"] = df_train_meta["age_approximate"]
df_train["sex"] = df_train_meta["sex"]

df_valid["image_id"] = "data_valid/"+df_valid["image_id"]
df_valid["valid"] = True
df_valid["age"] = df_valid_meta["age_approximate"]
df_valid["sex"] = df_valid_meta["sex"]

df = pd.concat([df_train, df_valid], ignore_index=True)
mapping = {0.0: "healthy", 1.0: "no healthy"}
df = df.replace({'melanoma': mapping, 'seborrheic_keratosis': mapping})
df.tail(10)
```

```python
data = ImageItemList.from_df(df, path, suffix=".jpg") # Step 1: Provide inputs
data = data.split_from_df(col='valid')                # Step 2: Split train/val
data = data.label_from_df("melanoma")                 # Step 3: Label the inputs
data = data.transform(get_transforms(), size=512)     # Step 4: Add transformation
data = data.databunch(bs=16)                          # Step 5: Convert to a DataBunch
data = data.normalize(imagenet_stats)                 # Step 6: Normalize data
data
```

```python
data.show_batch(rows=4, figsize=(10,6))
```

# Train

```python
arch   = models.resnet34 # models.resnet50

#acc_02 = partial(accuracy_thresh, thresh=0.2)
#f_score = partial(fbeta, thresh=0.2)

learn = create_cnn(data, arch, metrics=accuracy)
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(3, 1e-2)
```

```python
learn.recorder.plot_losses()
```

```python
learn.unfreeze()
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(3, 1e-5)
```

```python
learn.recorder.plot_losses()
```

# Results

```python
interp = ClassificationInterpretation.from_learner(learn)

losses,idxs = interp.top_losses()

len(data.valid_ds)==len(losses)==len(idxs)
```

```python
interp.plot_top_losses(9, figsize=(12,8))
```

```python
interp.plot_confusion_matrix()
```
