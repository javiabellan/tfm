```python
from fastai import *
from fastai.vision import *
```

# Data

```python
path     = pathlib.Path("D:/Datasets/TFM/histologyDS2828")
csv_path = path / "imageClasses.txt"
df       = pd.read_csv(csv_path, header=None, delim_whitespace=True, names=['Image', 'Label'])

mapping = {1: "conective", 2: "ephitelial", 3: "muscular", 4: "nervous"}
df = df.replace({'Label': mapping})

data = ImageDataBunch.from_df(path, df, folder='imgs', bs=16, ds_tfms=get_transforms(), size=224)


# normalising the dataset using the same normalisation applied to the imagenet dataset
data.normalize(imagenet_stats)


#data = ImageDataBunch.from_csv(path, csv_labels=train_lbl,folder='train', ds_tfms=tfms, size=sz, suffix='.tif',test=test_folder,bs=bs);
```

```python
print(data.classes)
len(data.classes),data.c
```

```python
data.show_batch(rows=4, figsize=(12,6))
```

# Training

```python
arch   = models.resnet34 # models.resnet50

#acc_02 = partial(accuracy_thresh, thresh=0.2)
#f_score = partial(fbeta, thresh=0.2)

learn = create_cnn(data, arch, metrics=accuracy)
```

```python
learn.recorder.plot_lr(show_moms=True)
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
1-0.145133
```

```python
learn.fit_one_cycle(3, 1e-2)
```

```python
learn.recorder.plot_lr(show_moms=True)
```

```python
learn.recorder.plot_losses()
```

```python
learn.unfreeze()
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(4, 1e-4)
```

```python
learn.recorder.plot_losses()
```

# Results

```python
interp = ClassificationInterpretation.from_learner(learn)

losses,idxs = interp.top_losses()

len(data.valid_ds)==len(losses)==len(idxs)
```

```python
interp.plot_top_losses(9, figsize=(12,8))
```

```python
interp.plot_confusion_matrix()
```
