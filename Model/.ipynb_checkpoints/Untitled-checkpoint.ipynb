```python
import matplotlib.pyplot as plt
from numpy import arange, sin, cos, pi
%pylab inline
pylab.rcParams['figure.figsize'] = (16, 5)
```

```python
t = arange(0.0, 10, 0.01)
line = [0] * 100

#cos(2*pi*t)
y  = [1/i * cos(i*t) for i in [1, 2, 3, 4, 5]]
y2 = [1/i * cos(i*t + (i-1)*pi/2) for i in [1, 2, 3, 4, 5]]
```

```python
plt.xlabel('time')
plt.ylabel('amplitude')
for i in [1, 2, 3, 4, 5]:
    plt.plot(t, y[i])
```

```python
y1_sum = [sum(x) for x in zip(*y)]
plt.xlabel('time')
plt.ylabel('amplitude')
plt.plot(t, y1_sum)
```

```python
plt.xlabel('time')
plt.ylabel('amplitude')
for i in [1, 2, 3, 4, 5]:
    plt.plot(t, y2[i])
```

```python
y2_sum = [sum(x) for x in zip(*y2)]
plt.xlabel('time')
plt.ylabel('amplitude')
plt.plot(t, y2_sum)
```
